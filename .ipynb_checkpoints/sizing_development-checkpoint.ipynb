### 3. Data Used for Work: ANSUR II Dataset (https://www.openlab.psu.edu/ansur2/)

### Data Cleaning

It would be a good idea to keep only those column which are directly related to torso and arms. 

For that reason we filter columns by specific keywords.

```python
relevant_cols = [col for col in data_m.columns if ('sleeve' in col) or ('shoulder' in col) \
                 or ('arm' in col) or ('wrist' in col) or ('hand' in col) \
                 or ('chest' in col) or ('abdominal' in col) or ('acromion' in col)\
                 or ('acromial' in col) or ('neck' in col) or ('waist' in col)]
len(relevant_cols)
```

```python
data_cl_m = data_m[relevant_cols]
data_cl_m.head()
```

```python
data_cl_m.describe()
```

### Random Forest Feature Importance Calculation

Since out target feature is 'sleeve length spine wrist' it should be good idea to remove variables directly related to the target such as 'sleeve outseam' and 'forearm hand length'

```python
not_rel_cols = ['sleevelengthspinewrist', 'sleeveoutseam', 'forearmhandlength', 
                'shoulderelbowlength','acromionradialelength','forearmcenterofgriplength',
               ]
for feature in not_rel_cols:
    relevant_cols.remove(feature)
```

```python
mms = MinMaxScaler()
mms.fit(data_cl_m) 
df_sc_m = pd.DataFrame(mms.transform(data_cl_m),
                        columns = data_cl_m.columns)
```

```python
X = df_sc_m.drop(not_rel_cols, axis=1)
Y = df_sc_m["sleevelengthspinewrist"]
#names = relevant_cols
rf = RandomForestRegressor()
rf.fit(X, Y)
print("Features sorted by their score:")
for feature in sorted(zip(map(lambda x: round(x, 3), rf.feature_importances_), relevant_cols), 
             reverse=True):
    print(feature)
```
